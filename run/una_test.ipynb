```python
from __future__ import annotations

from dataclasses import dataclass
from typing import Any

@dataclass
class AuthContext:
    role: str
    authenticated: bool

@dataclass
class IntentCapsule:
    original_user_prompt: str
    normalized_intent: str
    role: str
    password_verified: bool
    allowed_scope: list[str]
```

```python
def authenticate_user(role: str, password: str, auth_cfg: dict[str, str]) -> AuthContext:
    role = role.strip().lower()

    if role not in {"admin", "general"}:
        raise RuntimeError(f"Invalid role: {role}")

    if role == "admin":
        expected = auth_cfg["admin_password"]
    else:
        expected = auth_cfg["general_password"]

    if password != expected:
        raise RuntimeError("Authentication failed")

    return AuthContext(role=role, authenticated=True)
```
